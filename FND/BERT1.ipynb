```python
# Install required libraries
!pip install transformers
!pip install tensorflow

import pandas as pd
import numpy as np
import tensorflow as tf
from transformers import BertTokenizer, TFBertModel
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Tokenize input data
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def tokenize_data(texts, tokenizer, max_length=128):
    input_ids = []
    attention_masks = []

    for text in texts:
        encoded = tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=max_length,
            pad_to_max_length=True,
            return_attention_mask=True,
            return_tensors='tf',
            truncation=True
        )

        input_ids.append(encoded['input_ids'])
        attention_masks.append(encoded['attention_mask'])

    input_ids = tf.concat(input_ids, axis=0)
    attention_masks = tf.concat(attention_masks, axis=0)

    return input_ids, attention_masks

df = data_merge
input_ids, attention_masks = tokenize_data(df['text'].values, tokenizer)
```

```python
# Load the pre-trained BERT model for embedding
from transformers import BertTokenizer, TFBertModel

bert_model = TFBertModel.from_pretrained('bert-base-uncased')

def get_embeddings(input_ids, attention_masks, bert_model):
    outputs = bert_model(input_ids, attention_mask=attention_masks)
    pooled_output = outputs.pooler_output
    return pooled_output

embeddings = get_embeddings(input_ids, attention_masks, bert_model)
```

```python
# ACO optimization for feature selection
class ACOFeatureSelector:
    def __init__(self, n_ants=10, n_iterations=50):
        self.n_ants = n_ants
        self.n_iterations = n_iterations

    def select_features(self, embeddings):
        # Placeholder for ACO implementation
        # For now, selecting first 4 features as an example
        selected_features = embeddings[:, :4]
        return selected_features

aco_selector = ACOFeatureSelector()
selected_features = aco_selector.select_features(embeddings)
```

```python
# Prepare data for training and testing
X = selected_features.numpy()
y = df['class'].values

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Build a simple classifier on top of BERT embeddings
input_layer = tf.keras.layers.Input(shape=(4,), dtype=tf.float32)
dense_layer = tf.keras.layers.Dense(128, activation='relu')(input_layer)
output_layer = tf.keras.layers.Dense(len(np.unique(y)), activation='softmax')(dense_layer)

classifier_model = tf.keras.Model(inputs=input_layer, outputs=output_layer)

classifier_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the classifier
classifier_model.fit(X_train, y_train, epochs=3, batch_size=32, validation_split=0.2)

# Evaluate the classifier
loss, accuracy = classifier_model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}")
print(f"Test Accuracy: {accuracy}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data for training and testing
X = embeddings.numpy()
y = df['class'].values

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Build a simple classifier on top of BERT embeddings
input_layer = tf.keras.layers.Input(shape=(X.shape[1],), dtype=tf.float32)
dense_layer = tf.keras.layers.Dense(128, activation='relu')(input_layer)
output_layer = tf.keras.layers.Dense(len(np.unique(y)), activation='softmax')(dense_layer)

classifier_model = tf.keras.Model(inputs=input_layer, outputs=output_layer)

classifier_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the classifier
classifier_model.fit(X_train, y_train, epochs=3, batch_size=32, validation_split=0.2)

# Predict on test set
y_pred = classifier_model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Decode labels
y_test_decoded = label_encoder.inverse_transform(y_test)
y_pred_decoded = label_encoder.inverse_transform(y_pred_classes)

# Print classification report
print(classification_report(y_test_decoded, y_pred_decoded))

# Plot confusion matrix
cm = confusion_matrix(y_test_decoded, y_pred_decoded)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
